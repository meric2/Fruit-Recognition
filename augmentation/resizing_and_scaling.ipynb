```python
import random
from PIL import Image
import matplotlib.pyplot as plt  # For visualization
import os

def random_resize_and_scale(image, size_range=(256, 512), scale_range=(0.7, 1.3)):
    """
    Resize and scale the image within specified ranges.

    Parameters:
    - image: PIL Image object
    - size_range: A tuple of (min_size, max_size) for resizing
    - scale_range: A tuple of (min_scale, max_scale) for scaling

    Returns:
    - Resized and scaled PIL Image object
    """

    # Choose a random size within the specified range
    new_size = random.randint(*size_range)

    # Resize the image
    resized_image = image.resize((new_size, new_size), Image.ANTIALIAS)

    # Choose a random scale factor within the specified range
    scale_factor = random.uniform(*scale_range)

    # Scale the image
    scaled_image = resized_image.resize((int(new_size * scale_factor), int(new_size * scale_factor)), Image.ANTIALIAS)

    return scaled_image

# Example Usage
image_path = "data/apple/apple_1.jpg"
original_image = Image.open(image_path)

# Apply random resizing and scaling
resized_scaled_image = random_resize_and_scale(original_image)

# Display the original and augmented images for comparison
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Resized and Scaled Image")
plt.imshow(resized_scaled_image)
plt.axis("off")

plt.show()
```

```python
def save_resized_scaled_images(folder, image_path, image_order, num_augmentations=5):
    # Create a new folder for resized and scaled images
    resizing_scaling_folder = os.path.join(folder, "resizing_scaling")
    os.makedirs(resizing_scaling_folder, exist_ok=True)

    # Load the original image
    original_image = Image.open(image_path)

    # Apply resizing and scaling augmentations and save the images
    for i in range(1, num_augmentations + 1):
        # Apply random resizing and scaling
        augmented_image = random_resize_and_scale(original_image)
        augmented_image = augmented_image.convert("RGB")

        # Save the augmented image
        augmented_image_path = os.path.join(resizing_scaling_folder, f"{image_order}_resized_scaled_{i}.jpg")
        augmented_image.save(augmented_image_path)
```

```python
fruits = ["apple", "banana", "carrot", "cherry", "coconut", "cucumber", "grapefruit", "kiwi",
    "melon", "pear", "pineapple", "plum", "tomato", "watermelon"]
```

```python
import os

def count_images_in_folder(folder_path, image_extensions=['.jpg', '.jpeg', '.png', '.gif', '.bmp']):
    image_count = 0

    # Check if the folder exists
    if os.path.exists(folder_path) and os.path.isdir(folder_path):
        # List all files in the folder
        files = os.listdir(folder_path)

        # Count the image files
        for file in files:
            if any(file.lower().endswith(ext) for ext in image_extensions):
                image_count += 1
    else:
        print(f"The folder {folder_path} does not exist.")

    return image_count
```

```python
for fruit in fruits:
    # Count the number of images in the folder
    image_count = count_images_in_folder(f"data/{fruit}")
    for i in range(1, image_count + 1):
        # Create resized and scaled images
        folder_path = f"data/{fruit}"
        image_path = os.path.join(folder_path, f"{fruit}_{i}.jpg")
        save_resized_scaled_images(folder_path, image_path, f"{fruit}_{i}")
```
